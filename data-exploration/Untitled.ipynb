```python
import pandas as pd
import numpy as np
import os,glob

columns = ['DGUID', 'ALT_GEO_CODE', 'GEO_LEVEL', 'GEO_NAME', 'NEW_CHARACTERISTIC_NAME', 'C2_COUNT_MEN+', 'C3_COUNT_WOMEN+']
new_columns = {'DGUID': 'dguid', 'ALT_GEO_CODE':  'alt_geo_code' ,'GEO_NAME': 'geo_name', 'GEO_LEVEL': 'geo_level', 'NEW_CHARACTERISTIC_NAME': 'characteristic_name',
             'C2_COUNT_MEN+': 'men',  'C3_COUNT_WOMEN+': 'women'}

DATA_CSV = '/Users/boristsao/Documents/Projects/canada-vis/data-exploration/cleaned-data/age_groups_of_the_population/age_groups_of_the_population_Territories.csv'
```

```python
df = pd.read_csv(DATA_CSV)
df = df.fillna(0)
df.head()
```

```python
df = df[columns]
df = df.rename(columns=new_columns)
```

```python
df = df.melt(id_vars=['dguid', 'geo_name', 'alt_geo_code', 'geo_level', 'characteristic_name'], var_name='gender', value_name='values')
```

```python
df.to_csv('test.csv')
```

```python
csv_files = '/Users/boristsao/Documents/Projects/canada-vis/data-exploration/cleaned-data/age_groups_of_the_population/'
li = []
for csv_file in glob.glob(os.path.join(csv_files, '*.csv')):
        li.append(pd.read_csv(csv_file))
df = pd.concat(li, axis=0, ignore_index=True)
df = df.fillna(0)
df.head()
```

```python
df = df[columns]
df = df.rename(columns=new_columns)
df = df.melt(id_vars=['dguid', 'geo_name', 'alt_geo_code', 'geo_level', 'characteristic_name'], var_name='gender', value_name='values')
df.head()
```

```python
df.to_csv('test.csv', index=False)
```
