```python
# Loading the Libraries
from dotenv import load_dotenv
load_dotenv()
import os
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import os,glob
```

```python
# Change the column names
csv = '/Users/boristsao/Documents/Projects/canada-vis/data-exploration/final-data/age_groups_of_the_population.csv'
df = pd.read_csv(csv, encoding="ISO-8859-1",low_memory=False)
```

```python
data_path = '/Users/boristsao/Documents/Projects/canada-vis/data-exploration/final-data/'
for csv_file in glob.glob(os.path.join(data_path, '*.csv')):
    df = pd.read_csv(csv_file, encoding="ISO-8859-1",low_memory=False)
    new_columns = {}

    for column in df.columns:
        new_columns[column] = column.lower().replace(" ", "_")

    new_df = df.rename(columns = new_columns)
    new_df = new_df.fillna(0)
    new_df.to_csv(csv_file)
    
```

```python
new_columns = {}

for column in df.columns:
    new_columns[column] = column.lower().replace(" ", "_")

new_df = df.rename(columns = new_columns)
new_df = new_df.fillna(0)
new_df.head()
```

```python
ssl_args = {'ssl_ca': "/etc/ssl/cert.pem"}

conn_string = 'mysql+pymysql://' + os.getenv("USERNAME") + ':' + os.getenv("PASSWORD") + '@' + os.getenv("HOST") + '/' + os.getenv("DATABASE") 

engine = create_engine(conn_string, connect_args=ssl_args)
with engine.begin() as connection:
    new_df.to_sql('age_groups', con=connection, if_exists='append')
```
